#        Data Visualisation

```python
#plotting a chart between Quater, Year and Amount
sns.boxplot(x='Quater', y='Amount', hue='Year',  data=data)

#defining the axis
plt.xlabel('Quater')
plt.ylabel('Amount')
plt.title('Quater vs Amount with Year')
```

```python
#plotting a chart between Quater, Year and Amount
sns.violinplot(x='Sector', y='Amount', hue='Year',  data=data)

#defining the axis
plt.xlabel('Sector')
plt.ylabel('Amount')
plt.title('Sector vs Amount with Year')
```
